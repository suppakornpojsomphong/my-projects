```python
import csv
import json
from datetime import datetime

# Read CSV and convert each row into the desired JSON format
def csv_to_json(csv_file_path, json_file_path):
    with open(csv_file_path, mode='r', encoding='utf-8') as csv_file:
        csv_reader = csv.DictReader(csv_file)

        # List to hold the JSON objects
        json_data = []

        for row in csv_reader:
            # Create a dictionary for each row, formatting to match the provided JSON format
            json_object = {
                "name": row["Name"],
                "picture_url": row["Picture URL"],
                "description": row["Description"],
                "first_appearance": row["First Appearance"],
                "creator_name": row["Creator Name"],
                "species": row["Species"],
                "gender": row["Gender"],
            }
            # Append formatted row to json_data list
            json_data.append(json_object)

    # Write to a JSON file
    with open(json_file_path, mode='w', encoding='utf-8') as json_file:
        for item in json_data:
            json_file.write(json.dumps(item) + "\n")  # Write each JSON object on a new line

# Usage
csv_to_json('Sanrio_characters.csv', 'Sanrio.json')
```
